```python
#https://www.barttorvik.com/playerstat.php?link=y&sIndex=8&cvalue=ACC&year=2023&minmin=20&start=20221101&end=20230501
import pandas as pd
import random

df = pd.read_excel('ACC.xlsx', index_col=0)
South = ['Duke','Clemson','Florida St.','Georgia Tech','Wake Forest','Miami FL','North Carolina','North Carolina St.']

def returnStats(player):
    for i in range(1,len(df)+1):
        if player in df["PLAYER"][i]:
            allInfo = [df["TEAM"][i],df["YEAR"][i],df["MIN%"][i],df["TS"][i],df["AST"][i],df["TO"][i],df["REB"][i],df["PTS"][i]]
            return allInfo
    return []

def genHiddenPlayer():
    i = random.randint(1,len(df)+1)
    allInfo = [df["TEAM"][i],df["YEAR"][i],df["MIN%"][i],df["TS"][i],df["AST"][i],df["TO"][i],df["REB"][i],df["PTS"][i]]
    playerName = df["PLAYER"][i]
    return playerName, allInfo

def hint(hiddenPlayer,allInfo,playerGuessed):
    pgInfo = returnStats(playerGuessed)
    if len(pgInfo) == 0:
        return []
    allHint = []
    if playerGuessed in hiddenPlayer:
        print("Correct Player Guessed! You win!")
        return []
    for i in range(len(allInfo)):
        if i > 1:
            if allInfo[i] == pgInfo[i]:
                allHint.append("\u001b[32m\"Correct\"\u001b[0m")
            elif ((allInfo[i] - pgInfo[i])/allInfo[i]) < .5 and ((allInfo[i] - pgInfo[i])/allInfo[i]) > 0:
                allHint.append("\u001b[33m\"Too Low\"\u001b[0m")
            elif ((allInfo[i] - pgInfo[i])/allInfo[i]) > -.5 and ((allInfo[i] - pgInfo[i])/allInfo[i]) < 0:
                allHint.append("\u001b[33m\"Too High\"\u001b[0m")
            else:
                allHint.append("\x1b[31m\"Incorrect\"\x1b[0m")
        else:
            if allInfo[i] == pgInfo[i]:
                allHint.append("\u001b[32m\"Correct\"\u001b[0m")
            else:
                allHint.append("\x1b[31m\"Incorrect\"\x1b[0m")
    return allHint

def game():
    context = ["Team Name: ","Player Year: ","Minutes %: ","True Shooting: ","Assist %: ","Turnover %: ","Rebounds per game: ","Points per game: "]
    hiddenPlayer, allInfo = genHiddenPlayer()
    #print(hiddenPlayer)
    rd = 0
    print("Input 'Q' to quit.")
    while rd < 8:
        print("Round ",rd + 1)
        playerGuessed = input("Enter Name to Guess: ")
        if playerGuessed == "Q":
            print("Game over!")
            return 0
        allHint = hint(hiddenPlayer,allInfo,playerGuessed)
        pgInfo = returnStats(playerGuessed)
        if len(allHint) == 0:
            if playerGuessed in hiddenPlayer:
                return 0
            print("Invalid Player. Pick again.")
        if len(allHint) > 0:
            print("--------------------------")
            if (pgInfo[0] in South) == (allInfo[0] in South):
                print("Southern: " + str(pgInfo[0] in South) + ": " + "\u001b[32m\"Correct\"\u001b[0m")
            else:
                print("Southern: " + str(pgInfo[0] in South) + ": " + "\x1b[31m\"Incorrect\"\x1b[0m")
            for i in range(8):
                print(context[i] + str(pgInfo[i]) + ": " + allHint[i])
            print("--------------------------")
            rd = rd + 1
    print("You lost! Player was: ", hiddenPlayer)
```

```python
#Click me to play
game()
```
